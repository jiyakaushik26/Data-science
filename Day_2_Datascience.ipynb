List can contain anything-> number, set, dictionary , char, string and list itself


List takes time to execute


```python
lst = [1, 2, 3, "jiya", "kaushik", 'c', [2,4]]
print (lst)
```

Adding elements in a list-> use list.append

```python
lst = [1, 2, 3, "jiya", "kaushik", 'c', [2,4]]
lst.append('hello')
print (lst)

lst.append([12,13,14])
print (lst)
```

Extend basically adds the seperate element from the list

```python
lst = [1, 2, 3, "jiya", "kaushik", 'c', [2,4]]
lst.extend([12,13,14])
print (lst)
```

To insert the element in between-> use list.insert

the format is list.insert(index, value)

```python
lst = [1, 2, 3, "jiya", "kaushik", 'c', [2,4]]
# lst.insert(index,value)

lst.insert(2, "value")
print(lst)
```

in list extending only use those data structures which are iterable

in dictionary only keys are iterable so it will add only key

```python
lst2=["hi", "jiya", 3, [9,8,10]]

# to extend at last->
lst2.extend({'akshita':78})
print(lst2)

# to add at last->
lst2.append({5,6})
print(lst2)

#to insert in between->
lst2.insert(4,{'sun':3})
print(lst2)
```

pop-> it will remove the last element

```python
lst = [1, 2, 3, "jiya", "kaushik", 'c', [2,4]]
lst.pop()
print(lst)
```

remove-> it wil remove the first occurrence only

```python
lst = [1, 2, 3, "jiya", 2, "kaushik", 'c', [2,4]]
lst.remove(2)
print(lst)
```

Index-> it returns the index of the value and prints it

```python
lst = [1, 2, 3, "jiya", 2, "kaushik", 'c', [2,4]]
print(lst.index(2))
```

```python
lst = [1, 2, 3, "jiya", 2, "kaushik", 'c', [2,4]]
print(lst.count(2))
```

Sort-> use list.sort and it does not return anything so use print

```python
lst = [23,100,4,67,12]
lst.sort()
print(lst)
```

**Sorted**->returns the new sorted list so need to store it in variable always

```python
lst = [23,100,4,67,12]
a= sorted(lst)
print(a)
type(a)
```

Sum, min, max and sorted are not list attributes, they are proper functions so use them as function(list)  not  list.function

**Tuples**

DATA STSRUCTURES USED TO STORE VALUES

They are **IMMUTABLE**


```python
tp=(1,2,3,4)
print(tp)
```

since we cannot change or modify tuples there is no add, append, remove, pop functions

We cannot use sort as well because it will change the order

it will try to change in the same tuple which tuple wont allow

Sorted -> can be used in tuple because ut returns a new list

it will not modify anything in the tuple

```python
tp=(5,7,1,2)
sorted(tp)
```

sorted function returns a list

if you need to again get a sorted tuple use tuple word

```python
tp=(5,7,1,2)
a = tuple(sorted(tp))
print (a)
```

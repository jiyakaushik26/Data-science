**numpy.arange**-->is a built-in NumPy function that creates a one-dimensional array of evenly spaced numeric values within a defined interval

**Format->** **numpy.arange(start, stop, step, dtype=int)**

**stop** is exclusive.


```python
b = np.arange(3, 12, 4, dtype=int)
print(b)
```

Since only one argument is given, NumPy treats it as the stop value and by default step is 1 and start value is 0

Think of np.arange() exactly like Python's range(), but it returns a NumPy array instead of a range object.

```python
a = np.arange(5)
print(a)
```

**with endpoint=False->**  Now stop is not included.

# Broadcasting

It is NumPy's way of making arrays with different shapes work together in arithmetic operations.

```python
a = np.array([1, 2, 3])

print(a + 5)
```

```python
a = np.ones((3,3), dtype=int)

b = np.array([1,2,3])

print(a + b)
```

For broadcasting, dimensions must be:

Equal, OR

One of them must be 1

```python
a = np.ones((3,3), dtype = int)
b = np.ones((3,3), dtype = int)
print (a)

print(a*2)

print(a+b)

print(a-b)

print(a*b)

print(a//b)
# elements at corresponding positionn are being operated.

c=np.ones((2,2), dtype=int)
# print(a+c) -> ValueError : operands could not be broadcast together with different shapes
```

The **split()** method in Python breaks a single string into a list of smaller substrings based on a specified delimiter

**map(function, iterable)** applies the function to every element.

**map()** does not directly return a list. It returns a map object. so write list to convert it into a list

**np.argmax(scores)** -> Returns the index of the maximum value.

**np.argmin(scores)** -> Returns the index of the minimum value.

```python
scores = input()
# scores = np.random.randint(0,100,7)
scores = scores.split()
# scores =[int(x) for x in scores] -> another way to convert in integer
scores = list(map(int,scores))
days = ["MON","TUES","WED","THURS","FRI","SAT","SUN"]
scores = np.array(scores)
for i in range (len(scores)):
  print(scores[i],days[i])
print("maximum score : ", np.max(scores), "on ", days[np.argmax(scores)])
print("minimum score : ", np.min(scores), "on ", days[np.argmin(scores)])
print("sum of scores : ", np.sum(scores))
print("mean of scores : ", np.mean(scores))
print("standard deviation of scores", np.std(scores))
```

**Axis** can be used for any operation but the suntax would be

**np.func(a, axis=0)** where func can be sum,min,max,etc



**To count->** np.count_nonzero(condition, axis)

or

**To count->** np.sum(np.where(condition, 1, 0))

```python
a=np.random.randint(1,90, (5,5))
print(a)
b=np.mean(a,axis=1)
print(b)
c=np.mean(a,axis=0)
print(c)
d=np.where(a%2==0, -1, a)
print(d)
e=np.sort(a, axis=0)
print(e)
f=np.sort(a,axis=0)
print(f)
g=np.sum(a,axis=0)
print(g)
h=np.sum(a,axis=1)
print(h)
i = np.count_nonzero(a<0, axis=0)
print(i)

j=np.sum(np.where (a<0, 1, 0))
print(j)
```

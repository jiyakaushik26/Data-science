```python
s= ' today '
b= '    weather   '
c= 'is GooD      '

s=s.strip()
a = s.capitalize()

b=b.strip()
c=c.strip()

m= c.lower()
print(a+', '+b+' '+m)

result = a + ', ' + b + ' ' + m
result=result.replace('good', 'bad')
print(result)

r=result.find('bad')
print(r)
s=result.find('good')
print(s)

s1=result.startswith('Today')
print(s1)
s2=result.endswith('good')
print(s2)
s3=result.endswith('bad')
print(s3)
```

# **Dictionary**

A Python dictionary is a built-in data structure used to store data in key-value pairs.


1.   Keys must be unique:
2.   Keys must be immutable

1.   Values can be anything and even duplicate
2.   List item


 It is ** mutable (changeable)** and preserves insertion order.



** Accessing Values**-> Access values by passing the key inside square brackets [] or by .get() method to avoid errors if the key does not exist

```python
dct={
    'name': 'jiya',
    'age':23,
    'occupation': 'student',
    'class': 10,
    'roll-no': 18

}
print(dct)
```

```python
dct={
    'name': 'jiya',
    'age':23,
    'occupation': 'student',
    'class': 10,
    'roll-no': 18

}
print(dct['name'])
print(dct['name'][0])
```

**To add a new value in dictionary**

Add new pairs or update existing ones using the assignment operator

```python
dct={
    'name': 'jiya',
    'age':23,
    'occupation': 'student',
    'class': 10,
    'roll-no': 18

}
dct['id']=6435
print(dct)
```

To **delete** any key and its value

```python
dct={
    'name': 'jiya',
    'age':23,
    'occupation': 'student',
    'class': 10,
    'roll-no': 18,
    'id':6789

}
del dct['id']
print(dct)
```

A Pandas **DataFrame** is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns)

df is a DataFrame object

**pd.DataFrame(...)** is a Pandas function that converts data into a table.

display(df)-> Shows a prettier table with proper row/column formatting.

we can also write simply df to print the data

pd.DataFrame(file) = convert file → table

```python
dct={
    'name':['Ton', 'Krish', 'Vani', 'Jiya'],
    'age':[20,11,18,19],
    'roll': [1,2,3,4]
}
df=pd.DataFrame(dct)
df

display(df)
```

To get filtered data->

use **df.loc(what you want)**

df.loc is a Pandas accessor used to select data from a DataFrame by row labels and column names.

```python
df=pd.read_csv('student_data.csv')
df.loc[df['study_hours']==11]
```

**df.loc[df['study_hours']==11, ['study_hours', 'sleep_hours']].iloc[4:8]**->

"Find all students whose study_hours equals 11, keep only the study_hours and sleep_hours columns, then show the 5th through 8th matching rows."

**iloc** access data through index only

**loc** access data through labels

iloc apni indexing or kaam krta hai

```python
df=pd.read_csv('student_data.csv')
df.loc[df['study_hours']==11, ['study_hours', 'sleep_hours']].iloc[4:8]

```
